```python
import pandas as pd
import numpy as np
```

```python
dija = pd.DataFrame.from_csv('DJIA.csv', index_col='DATE').sort()
debt = pd.read_csv('debt_google_trend.csv', index_col='date')
dija.index = dija.index.to_datetime()
debt.index = debt.index.to_datetime()
```

```python
dija.rename(columns={'VALUE': 'dija'}, inplace=True)
```

```python
data = dija.join(debt.tshift(1, freq='D')).dropna()
```

```python
data.head()
```

```python
data.dija.plot()
```

```python
data['log_price'] = np.log(data.dija)
data['rolling_mean'] = pd.rolling_mean(data.debt, 5)
```

```python
data['signal_bias'] = data.debt > data.rolling_mean.shift(-1)
data['order_bias'] = 1
data['order_bias'][data.signal_bias == False] = -1

data['signal'] = data.debt.shift(-1) > data.rolling_mean.shift(-2)
data['order'] = 1
data['order'][data.signal == False] = -1
```

```python
(-1*data.order*data.log_price + data.order*data.log_price.shift(1)).cumsum().plot();
```

```python
(-1*data.order_bias*data.log_price + data.order_bias*data.log_price.shift(1)).cumsum().plot();
```
