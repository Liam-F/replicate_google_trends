```python
import pandas as pd
import numpy as np
```

```python
dija = pd.DataFrame.from_csv('DJIA.csv', index_col='DATE').sort()
debt = pd.read_csv('debt_google_trend.csv', index_col='date')
dija.index = dija.index.to_datetime()
debt.index = debt.index.to_datetime()
debt.dtypes
```

```python
debt = debt.tshift(2, freq='D') # Moves Saturday to Monday
debt.head()
```

```python
dija.rename(columns={'VALUE': 'dija'}, inplace=True)
dija.head()
```

```python
data = dija.join(debt).dropna()
```

```python

data.head()
```

```python
data.dija.plot()
```

```python
data.debt.plot()
```

```python
data['rolling_mean'] = pd.rolling_mean(data.debt, 5)
data.head()
```

```python
data['signal_bias'] = data.debt > data.rolling_mean.shift(1)
data['order_bias'] = 1
data['order_bias'][data.signal_bias == False] = -1

data['signal'] = data.debt.shift(1) > data.rolling_mean.shift(2)
data['order'] = 1
data['order'][data.signal == False] = -1
data['order'].ix[:5] = 0

from IPython.display import HTML
HTML(data.head(20).to_html())
```

```python
initial_capital = 1000000 # US Dollars
single_trade_capital = 100000 # US Dollars
data['shares_traded'] = single_trade_capital / data.dija
data['pnl'] = - data.dija * data.shares_traded * data.order + data.dija.shift(1) * data.shares_traded * data.order
HTML(data.head(20).to_html())
```

```python
bankroll = initial_capital + data.pnl.cumsum()
bankroll.plot()
```
