```python
import pandas as pd
import pandas.io.data
import numpy as np

delta_t = 3
```

Load data: Dow Jones Industrial Average, S&P500 from yahoo and the google trend weekly averages.

```python
dija = pd.DataFrame.from_csv('DJIA.csv', index_col='DATE').sort()
sp500 = pd.DataFrame(pd.io.data.DataReader("^GSPC", "yahoo", start=datetime.datetime(2004, 1, 1), end=datetime.datetime(2011, 4, 1))['Adj Close'], columns=['sp500'])
debt = pd.read_csv('debt_google_trend.csv', index_col='date')
dija.index = dija.index.to_datetime()
debt.index = debt.index.to_datetime()
```

```python
debt = debt.tshift(2, freq='D') # Moves Saturday to Monday
debt.head()
```

```python
dija.rename(columns={'VALUE': 'dija'}, inplace=True)
dija.head()
```

```python
# Join together into single dataframe.
data = pd.concat([dija, sp500, debt], axis=1).dropna()
data.head()
```

```python
data['rolling_mean'] = pd.rolling_mean(data.debt, delta_t)
data.head()
```

Generate the trading and order signals.
From the paper:
"We use Google Trends to determine how many searches n(t – 1) have been carried out for a specific search term such as debt in week t – 1, where Google defines weeks as ending on a Sunday, relative to the total number of searches carried out on Google during that time."
and
"We implement this strategy by selling the DJIA at the closing price p(t) on the first trading day of week t, if Δn(t − 1, Δt) > 0, and buying the DJIA at price p(t + 1) at the end of the first trading day of the following week. [...]. If instead Δn(t − 1, Δt) < 0, then we buy the DJIA at the closing price p(t) on the first trading day of week t and sell the DJIA at price p(t + 1) at the end of the first trading day of the coming week."

```python
data['order'] = 0
data['order'][data.debt.shift(1) > data.rolling_mean.shift(2)] = -1 # Shorting if cross-over
data['order'][data.debt.shift(1) < data.rolling_mean.shift(2)] = 1 # Buying otherwise.
data['order'].ix[:delta_t] = 0
```

Generate the same signals but with look-ahead bias where we are using the current week search-term. Note that this would in reality not be available.

```python
data['order_bias'] = 0
data['order_bias'][data.debt > data.rolling_mean.shift(1)] = -1 # Shorting if cross-over
data['order_bias'][data.debt < data.rolling_mean.shift(1)] = 1 # Buying otherwise.
data['order_bias'].ix[:delta_t] = 0
```

From the paper:
"If we take a ‘short position’—selling at the closing price p(t) and buying back at price p(t + 1)—then the cumulative return R changes by log(p(t)) − log(p(t + 1)). If we take a ‘long position’—buying at the closing price p(t) and selling at price p(t + 1)—then the cumulative return R changes by log(p(t + 1)) − log(p(t))."

```python
initial_capital = 1000000 # US Dollars
single_trade_capital = 100000 # US Dollars
data['shares_traded'] = single_trade_capital / data.dija
# I think the signs should actually be reversed here...
data['pnl'] = data.dija * data.shares_traded * data.order - data.dija.shift(1) * data.shares_traded * data.order
data['log_returns'] = data.order * np.log(data.dija) - data.order * np.log(data.dija.shift(1))
data['log_returns_sp'] = data.order * np.log(data.sp500) - data.order * np.log(data.sp500.shift(1))
data['log_returns_bias'] = data.order_bias * np.log(data.dija) - data.order_bias * np.log(data.dija.shift(1))

HTML(data.head(20).to_html())
```
